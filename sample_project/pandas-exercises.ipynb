```python
# Import the car sales DataFrame with missing data
url = "https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-missing-data.csv"
car_sales_missing = pd.read_csv(url)
car_sales_missing
```

```python
# Fill the Odometer (KM) column missing values with the mean of the column inplace
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
car_sales_missing
```

```python
# remove the rest of the missing data in place
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
# Create a "Seats" column where every row has a value of 5
seats_column = pd.Series([5 for x in range(6)])
car_sales_missing["Seats"] = seats_column
car_sales_missing
```

```python
import random
# Create a column called "Engine Size" with random values between 1.3 and 3.5
engine_size_column = pd.Series([1.3 + random.random() * (3.5 - 1.3) for x in range(6) ])
car_sales_missing["Engine Size"] = engine_size_column
car_sales_missing
```

```python
car_sales_missing["Price"] = car_sales_missing["Price"].str.replace('[\$\,\.]', '').astype(int)
car_sales_missing
```

```python
# Create a column which represents the price of a car per kilometer
car_sales_missing["Price per KM"] = car_sales_missing["Price"]  / car_sales_missing["Odometer"]
car_sales_missing
```

```python
# remove Price per KM using drop()
car_sales_missing = car_sales_missing.drop("Price per KM", axis=1)
car_sales_missing
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the shuffled DataFrame to a new variable
car_sales_shuffled = car_sales_missing.sample(frac=1)
car_sales_shuffled
```

```python
# reset the indexes of the shuffled DataFrame
car_sales_shuffled.reset_index(inplace=True, drop=True)
car_sales_shuffled
```

```python
# Change the Odometer values from kilometers to miles using a Lambda function
# Then view the DataFrame
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].apply(lambda x: x / 1.6)
car_sales_missing
```
