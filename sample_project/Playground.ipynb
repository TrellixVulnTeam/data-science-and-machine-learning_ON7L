```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url = "https://www.quandl.com/api/v3/datasets/USTREASURY/REALYIELD.csv?api_key=tj-rs-5cAqE_d7Qmo9_s"
treasury_data = pd.read_csv(url)

treasury_data.head()
```

```python
del treasury_data["7 YR"]
del treasury_data["20 YR"]

treasury_data.head()
```

```python
treasury_data["5 YR"] = treasury_data["5 YR"].astype(float) / 100
treasury_data["10 YR"] = treasury_data["10 YR"].astype(float) / 100
treasury_data["30 YR"] = treasury_data["30 YR"].astype(float) / 100
treasury_data['Date'] = pd.to_datetime(treasury_data['Date'])

treasury_data.head()
```

```python
date = treasury_data["Date"]
five = treasury_data["5 YR"]
ten = treasury_data["10 YR"]
thirty = treasury_data["30 YR"]

plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(date, five, label="5 Yr")
ax.plot(date, ten, label="10 Yr")
ax.plot(date, thirty, label="30 Yr")

ax.legend(loc="lower left")

ax.set_title("U.S. Real Yields", fontsize=16, fontweight="bold")
ax.set(ylabel="Real Yield")
ax.axhline(0, linestyle='--', color="red")

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.1%}'.format(x) for x in vals])

plt.show()
```
